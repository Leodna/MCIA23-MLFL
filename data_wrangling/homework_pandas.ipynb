```python
import numpy as np
import pandas as pd
import random
```

```python
#Read the data
df = pd.read_csv('data/NFL.csv')
df1 = pd.read_csv('data/NFL.csv')
df2 = pd.read_csv('data/NFL.csv')

df.head()
```

```python
def weiValCol(col,df):
    #Get element wieght, element_weight = the time that the element occurs
    valCounts = df[col].value_counts()
    #Get the sum of all the weights
    total = valCounts.sum()
    #The times that a value occurs divided by the total number of values to get the probability
    weights = valCounts/total        
    #Get the unique values of column,NaN val is in this list, drop that value, and put into a list
    valList = df[col].dropna().unique().tolist()                
    return valList,weights
```

```python
#Get the columns wich have NaN values
nanCols = df.isna().any()
#Pandas Seires to list
nanCols = df.columns[nanCols].tolist()
#Show the columns wich have NaN values
print (nanCols)
```

```python
#Loop trhough the Columns with NaN values
for c in nanCols:    
    #Conditional to know if is numerical or not
    if df[c].dtype == np.dtype(float) or df[c].dtype == np.dtype(int):
        #df1 fill missing numerical values with random number betwen the min and max value of the column        
        df1[c].fillna(pd.Series(np.random.uniform(df1[c].min(), df1[c].max(), size=len(df1))), inplace=True)        
        #df2 fill missing numerical values with the mean of the column
        df2[c].fillna(df2[c].mean(),inplace=True)

    else:
        #df1 fill missign categorical values with the most prominent category
        df1[c].fillna(df1[c].value_counts().index[0],inplace=True)
        #df2 fill missing categorical values with weighted random choice
        valList,weights = weiValCol(c,df2)
        #df2[c].fillna(random.choices(valList,weights=weights)[0],inplace=True)
        df2[c].fillna(np.random.choice(valList,1,p=weights)[0],inplace=True)
```
