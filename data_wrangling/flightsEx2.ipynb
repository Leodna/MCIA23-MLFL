```python
import pandas as pd
import numpy as np
```

```python
#Read the data
df = pd.read_csv('./data/nycflights13.csv',sep=',')
df.head()
```

```python
df1 = df.rename(columns={   'dep_time':'departure_time',
                            'arr_time':'arrive_time',
                            'arr_delay':'arrive_delay',
                            'dest':'destination'})
df1.columns
```

```python
#Filtering dataset
#'BOS' will be the airport
airportFilterVal = 'BOS'
#Making the filter if destination is BOS or if the origin is BOS
airportFilter = (df1['destination']==airportFilterVal) | (df1['origin']==airportFilterVal)
#Using the filter in the dataset
df2 = df1[airportFilter]
#Show the last 10 rows of the dataset filtered
df2.tail(10)
```

```python
#Create hierarchical indexing with the Outgoing and the Incoming flight categories
df3 = df2.set_index(['origin','destination'])
df3.head(10)
```

```python
#Sorting the flights on delay time in Ascending order (from smallest to greatest)
df4 = df1.copy()
#It add an extra column with the total delay
df4['total_delay'] = df4['departure_time'] + df4['arrive_delay']
df4 = df4.sort_values(by='total_delay',ascending=True)
df4.head()
```

```python
#Reindex the rows based on this new order so the index
df4.reset_index()
df5 = df4.reset_index().drop('index',axis=1)
df5.head()
```
